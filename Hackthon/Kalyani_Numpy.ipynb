```python
#9.Get the list of patient ID's which are not there in ReadmissionRegistry.

#loading Patients Dataset
df = pd.read_excel("HospitalDatabase .xlsx",sheet_name="Patients")
#Creating new dataframe with unique patient ids in patients dataset
DF_P= df['PatientID'].unique()

#Loading ReadmissionRegistry Dataset
df = pd.read_excel("HospitalDatabase .xlsx",sheet_name="ReAdmissionRegistry")
#Creating new dataframe with unique patient ids in ReAdmissionRegistry dataset
DF_RARP = df['PatientID'].unique()

#Getting Patients list

def non_match_Patients(DF_P, DF_RARP):
    non_match = []
    for i in DF_P:
        if i not in DF_RARP:
            non_match.append(i)
    return non_match

non_match = non_match_Patients(DF_P, DF_RARP)
print("list of patient ID's which are not there in ReadmissionRegistry: ", non_match)

print("Total Patients which are not there in ReadmissionRegistry :",len(non_match))
```

```python
#23 write a function to find the center element in the list. (one element in odd length list,  two elements in even length list)
def findMiddle(input_list):
    middle = float(len(input_list))/2
    if ((middle % 2 )!= 0):
        return input_list[int(middle - .5)]
    else:
        return ([input_list[int(middle-1)], input_list[int(middle)]])


input_string= input("Enter a list element separated by space: ")
list  = input_string.split()
print(list)
findMiddle(list)
```

```python
#78. Find reasonForVisit with highest count of acuity 5 patients.
df1=pd.read_excel("HospitalDatabase .xlsx", sheet_name ='EDVisits')
df2=df1[df1['Acuity']==5]
df2 = df2[['ReasonForVisit','Acuity']]
df3 = df2.groupby('ReasonForVisit')['Acuity'].apply(lambda x: (x== 5).sum()).sort_values(ascending=False).head(1)
df3
```

```python
#79 Which PrimaryDiagnosis has the biggest difference between maximum and minimum Expected LOS?
df1=pd.read_excel("HospitalDatabase .xlsx", sheet_name ='Discharges')
res=((df1.groupby('PrimaryDiagnosis')['ExpectedLOS'].max()) - (df1.groupby('PrimaryDiagnosis')['ExpectedLOS'].min())).sort_values(ascending=False).head(1).reset_index(name ='LOS Difference')
res
```
